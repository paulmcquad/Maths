## Advanced Plotting

 - Algebra
 - Derivatives

```python
from numpy import *
from pylab import *

# example data with some peaks:
x = linspace(0,4,1e3)
data = .2*sin(10*x)+ exp(-abs(2-x)**2)
```

```python
firstDeriv = diff(data)
```

```python
plot(x[0:999], (firstDeriv));
plt.show()
```

```python
a = diff(sign(diff(data))).nonzero()[0] + 1 # local min+max
b = (diff(sign(diff(data))) > 0).nonzero()[0] + 1 # local min
c = (diff(sign(diff(data))) < 0).nonzero()[0] + 1 # local max
```

```python
print('a min-max',a)
print('b local min',b)
print('c local max',c)
```

```python
from pylab import *
plot(x,data)
plot(x[0:999], (firstDeriv), label='First Derivative')
plot(x[b], data[b], "o", label="b -- min")
plot(x[c], data[c], "o", label="c -- max")
legend()
show()
```
