```python
from sympy import plot_implicit,And,symbols,Eq,Or

X1,X2=symbols('X1 X2')

p0=plot_implicit(And((X1**2*X2/20>1),(80/(X1**2+8*X2+5)>1),((X1+X2-5)**2/30+(X1-X2-12)**2/120>1)),
              (X1,0,10),(X2,0,10),show=False,line_color='g')
p1=plot_implicit(Or(Eq(X1**2*X2/20,1),Eq(80/(X1**2+8*X2+5),1),Eq((X1+X2-5)**2/30+(X1-X2-12)**2/120,1)), (X1,0,10),(X2,0,10),show=False,line_color='k')
p0.extend(p1)
p0.show()
```
